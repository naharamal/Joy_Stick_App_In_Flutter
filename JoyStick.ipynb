```python
import threading
import requests
from sendkeys import PressKey,ReleaseKey
import time
import pyautogui as gui
import pydirectinput
```

```python
forward_event=False
backward_event=False
left_event=False
right_event=False
```

```python
def forward():
    global forward_event
    while(True):
            time.sleep(.5)
            while(forward_event):
                PressKey(0x39)
                time.sleep(0.05)
            ReleaseKey(0x39)
            forward_event=False
            
```

```python
def backward():
    global backward_event
    while(True):
            time.sleep(.5)
            while(backward_event):
                print(backward_event)
                PressKey(0xD0)
                time.sleep(0.05)
            ReleaseKey(0xD0)
            backward_event=False
```

```python
def right():
    global right_event
    while(True):
            time.sleep(.5)
            while(right_event):
                PressKey(0xCD)
                time.sleep(0.05)
            ReleaseKey(0xCD)
            right_event=False
        
```

```python
def left():
    global left_event
    while(True):
            time.sleep(.5)
            while(left_event):
                print(left_event)
                PressKey(0xCB)
                time.sleep(0.05)
            ReleaseKey(0xCB)
            left_event=False
```

```python
def mouse_control(control_input):
    position=control_input.split(" ")
    x=float(position[1][0:4])
    y=float(position[2][0:4])
    d=float(position[3])
    print(d)
    pydirectinput.moveRel(int(x*70),int(y*70),duration=(1/(d*3)))
```

```python
def convert_to_action(control_input):
        
        global forward_event
        global backward_event
        global left_event
        global right_event
        
        if(control_input[0]=="M"):
            print(1)
            mouse_control(control_input)
        else:

            if(control_input=="forward"):
                print(control_input)
                forward_event=True
                left_event=False
                backward_event=False
                right_event=False
                time.sleep(.5)

            if(control_input=="left"):
                print(control_input)
                left_event=True
                forward_event=False
                backward_event=False
                right_event=False
                time.sleep(.5)

            if(control_input=="right"):
                print(control_input)
                right_event=True
                forward_event=False
                left_event=False
                backward_event=False
                time.sleep(.5)
            if(control_input=="backward"):
                print(control_input)
                backward_event=True
                forward_event=False
                left_event=False
                right_event=False
                time.sleep(.5)
            if(control_input=="nothing"):
                backward_event=False
                forward_event=False
                left_event=False
                right_event=False                


        
```

```python
def control_input():
    global forward_event
    global backward_event
    global left_event
    global right_event
    while(True):
        try:
            input_control=requests.get("http://192.168.43.1:8080/")
            input_control=input_control.content.decode("utf-8")
            convert_to_action(input_control)
            if(input_control=="esc"):
                print(6)
                pydirectinput.press('esc')
        except:
            pass

    
```
